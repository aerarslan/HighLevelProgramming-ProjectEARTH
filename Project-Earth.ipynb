## Dataset

The original data set is structured as:

* The event's index appears in the first column.
* The index of the prior event that caused it appears in the second column (-1 if no ancestor is found).
* The third column shows the time in seconds since 0:00 on January 1, 1982, when the earthquakes occurred.
* The fourth column represents the earthquake's magnitude, which ranges from [0,8].
* The 3D coordinates (in meters) of the earthquake hypocenter, i.e. the place from whence it began, are found in columns 5, 6, and 7. Latitude, longitude, and depth are used to create these Euclidean coordinates.

```python
# Column names to be used while reading the data
column_names = ["IndexOfEvent", "IndexOfPreviousEvent", "Second", "Magnitude", "ECEF-X", "ECEF-Y","ECEF-Z"]

# Reading the data
data = pd.read_csv("SouthCalifornia-1982-2011.dat", sep = " ", names = column_names)

# Sort based on the seconds. It is not sorted by default!!
data = data.sort_values(by = ['Second'])

# Display the head of the data
display(data.head())

# Display descriptive statistics include those that summarize the central tendency, 
# dispersion and shape of a dataset’s distribution
display(data.describe())
```

#### New Column "Date"

We added an 8th column to the data called "Date" which shows the exact day of when the event occurred. This column later will be used while visualizing the earthquakes month by month.

```python
# datetime.fromtimestamp function takes 01.01.1970 as the reference point.
# Since our reference point is 01.01.1982, we need to set our data based on that.

original_date = datetime(1970, 1, 1)
ref_date = datetime(1982, 1, 1)
dif = (ref_date - original_date).total_seconds()

date = []
for data_sec in data['Second']:
    data_year = datetime.fromtimestamp(data_sec + dif).year
    data_month = datetime.fromtimestamp(data_sec + dif).month
    data_day = datetime.fromtimestamp(data_sec + dif).day
    date.append(datetime(data_year, data_month, data_day))
    
data['Date'] = date

# Display the head of the data
display(data)
```

#### New Columns "Lat, Lot, Alt"

We added a 9th, 10th, 11th column to the data called "Lat, Lot, Alt" which shows the latitude, longitude, and altitude of the earthquakes. These columns will be used while marking the earthquakes on a world map. Alt is also used while visualizing the altitude of the earthquakes. We used "transform" function from the "pyproj" library for the conversion.

```python
# ---------------  Some Useful links that are used in this task  ---------------
# https://dar.vin/q3ZvQ
# https://hal.archives-ouvertes.fr/hal-01704943v2/document
# https://en.wikipedia.org/wiki/Geographic_coordinate_conversion#The_application_of_Ferrari's_solution
# https://pyproj4.github.io/pyproj/stable/api/transformer.html
# https://gfzpublic.gfz-potsdam.de/rest/items/item_43361/component/file_816919/content
# https://github.com/pyproj4/pyproj/issues/109
# https://pyproj4.github.io/pyproj/stable/gotchas.html#upgrading-to-pyproj-2-from-pyproj-1
# https://stackoverflow.com/questions/30307311/python-pyproj-convert-ecef-to-lla
# https://en.wikipedia.org/wiki/Geographic_coordinate_conversion#From_geodetic_to_ECEF_coordinates
# https://en.wikipedia.org/wiki/Geographic_coordinate_system
# Definition of the two reference frames 
# https://www.oc.nps.edu/oc2902w/coord/llhxyz.htm
# ------------------------------------------------------------------------------

# reference
ecef = pyproj.Proj(proj = 'geocent', ellps = 'WGS84', datum = 'WGS84') #euclidean x/y/z
lla  = pyproj.Proj(proj = 'latlong', ellps = 'WGS84', datum = 'WGS84') #latlon lat/lon/alt

# temp lists to hold x,y,z of the rows
x_list = []
y_list = []
z_list = []

# converting x, y, z into lat, lon, alt
for i in range(len(data)):
    x = data['ECEF-X'][i]
    y = data['ECEF-Y'][i]
    z = data['ECEF-Z'][i]
    
    lat, lon, alt  = pyproj.transform(ecef, lla, x, y, z, radians = False)
    
    x_list.append(lat)
    y_list.append(lon)
    z_list.append(alt)

# adding lat, lon, alt to the data
data['Lat'] = x_list
data['Lon'] = y_list
data['Alt'] = z_list

# save to to be used later whenever needed to save time
data.to_pickle("data.pkl")

display(data.head())
```

According to the visualization of the earthquakes, we saw that earthquakes with large magnitude trigger lots of earthquakes with smaller magnitude and closer to each other according to where it happens. So to understand the relation of earthquakes far away from each other, we sorted the data according to the time of the events and do rest of our analyzes.

```python
data = pd.read_pickle("data.pkl")

distance_to_ancestor = []
waiting_time_ancestor = []

for i in range(len(data['IndexOfPreviousEvent'])):
    
    if data['IndexOfPreviousEvent'][i] != -1:
        
        lat_dif = data['ECEF-X'][i] - data['ECEF-X'][data['IndexOfPreviousEvent'][i]]
        lon_dif = data['ECEF-Y'][i] - data['ECEF-Y'][data['IndexOfPreviousEvent'][i]]
        alt_dif = data['ECEF-Z'][i] - data['ECEF-Z'][data['IndexOfPreviousEvent'][i]]
        
        distance = np.sqrt(lat_dif**2 + lon_dif**2 + alt_dif**2)
        distance_to_ancestor.append(distance)
        
        waiting_time = data['Second'][i] - data['Second'][data['IndexOfPreviousEvent'][i]]
        waiting_time_ancestor.append(waiting_time)
        
    else:
        distance_to_ancestor.append(0)
        waiting_time_ancestor.append(0)
        
data['Distance-to-Ancestor'] = distance_to_ancestor
ancestor_data = data[data['Distance-to-Ancestor'] > 0]
print("Minimum distance to the ancestor: " + str(ancestor_data['Distance-to-Ancestor'].min()) + " meters")
print("Maximum distance to the ancestor: " + str(ancestor_data['Distance-to-Ancestor'].max()) + " meters")
print("Mean distance to the ancestor: " + str(ancestor_data['Distance-to-Ancestor'].mean()) + " meters")
print("Number of earthquakes that is closer to the ancestor smaller than mean distance: " + 
      str(len(ancestor_data[ancestor_data['Distance-to-Ancestor'] <= ancestor_data['Distance-to-Ancestor'].mean()])))
print("Number of earthquakes that is closer to the ancestor bigger than mean distance: " + 
      str(len(ancestor_data[ancestor_data['Distance-to-Ancestor'] >= ancestor_data['Distance-to-Ancestor'].mean()])))


data['Waiting-Time-Ancestor'] = waiting_time_ancestor
ancestor_data = data[data['Waiting-Time-Ancestor'] > 0]
print("\nMinimum waiting time with the ancestor: " + str(ancestor_data['Waiting-Time-Ancestor'].min()) + " seconds")
print("Maximum waiting time with the ancestor: " + str(ancestor_data['Waiting-Time-Ancestor'].max()) + " seconds")
print("Mean waiting time the ancestor: " + str(ancestor_data['Waiting-Time-Ancestor'].mean()) + " seconds")
print("Number of earthquakes that is closer to the ancestor smaller than mean waiting time: " + 
      str(len(ancestor_data[ancestor_data['Waiting-Time-Ancestor'] <= ancestor_data['Waiting-Time-Ancestor'].mean()])))
print("Number of earthquakes that is closer to the ancestor bigger than waiting time: " + 
      str(len(ancestor_data[ancestor_data['Waiting-Time-Ancestor'] >= ancestor_data['Waiting-Time-Ancestor'].mean()])))
```

#### Finding waiting time between each earthquake based on their "second" column.

Here, first eliminated some earthquakes based on different values of $m$. Later we calculated the waiting time between each subsequent earthquake.

```python
# loading the data
data = pd.read_pickle("data.pkl")

# getting only "Magnitude" and "Second" columns. 
data_time_mag = data[['Magnitude', 'Second']].copy()

# magnitude list
magnitudes = [2, 3, 4, 5, 6]

# the function that creates dataframes that include waiting time based on the given magnitude.
def find_waiting_time(data, m):
    data = data[data['Magnitude'] >= m]
    data['WaitingTime'] = data['Second'].diff()
    return data

# a list that holds different data frames filtered by magnitude
different_m_waiting_time_list = []
for magnitude in magnitudes:
    different_m_waiting_time_list.append(find_waiting_time(data_time_mag, magnitude))
    
for i in range(len(magnitudes)):
    print("Number of Earthquakes equal or bigger than m " + str(magnitudes[i]) +" : " + str(len(different_m_waiting_time_list[i])))
```

#### Fitting the data
We fit the data via optimizing the f(x) function with using optimize.curve_fit function in Scipy. 

$$f(x)= a{x^{-1}} + b$$

```python
# getting the h, and bins of waiting times with "m = 3"
h, bins, patches = axs[0].hist(different_m_waiting_time_list[1]['WaitingTime'],
                            250, facecolor="blue", alpha=0.75, edgecolor = 'purple')

# calculating the centers
center = (bins[1:] + bins[:-1]) / 2

# clearing the noisy part
#center = center[5:250]
#h = h[5:250]

# fit function
def fun(x,a,b):
    return a * (x**(-1)) + b

# fit curve
popt, pcov = optimize.curve_fit(fun, center, h)

# parameters
a, b = popt

# plotting
fig, ax = plt.subplots(figsize=(8, 8))
fig.suptitle('Scattered - Frequency of waiting time with the fit', fontsize = 20)
ax.scatter(center, h, label="data", s=10, c="black", marker="o")
#ax.set_yscale('log') #for log-log scale
#ax.set_xscale('log') #for log-log scale
ax.set_xlabel('Waiting time between events(sec)', 
               fontweight ='bold', fontsize = 15)
ax.set_ylabel('Freq.', 
               fontweight ='bold', fontsize = 15)

# setting x, y
x_line = np.arange(min(center), max(center), 1)
y_line = fun(x_line, a, b)
#ax.set_ylim(-1000, -100000000) #for log-log scale
ax.text(0.95, 0.75, 'm = 3', horizontalalignment='right',
     verticalalignment='bottom', transform=ax.transAxes, bbox=props, fontsize = 15)

# create a line plot for the function
ax.plot(x_line, y_line, '--', color='red', label="fit")
ax.legend(prop={'size': 16})
plt.show()

print("best params: ", "a=",a," ","b=",b)
```

#### Finding the distance between each earthquake based on their "Second" column.

Here, first eliminated some earthquakes based on different values of $m$. Later we calculated the distance between each subsequent earthquake.

```python
# loading the data
data = pd.read_pickle("data.pkl")

# getting only "Magnitude" and "Second" columns. 
data_distance_mag = data[['Magnitude', 'Second', 'ECEF-X', 'ECEF-Y', 'ECEF-Z']].copy()

# magnitude list
magnitudes = [2, 3, 4, 5, 6]

# the function that creates dataframes that include distance based on the given magnitude.
def find_distance(data, m):
    data = data[data['Magnitude'] >= m]
    lat_dif = data['ECEF-X'].diff()
    lon_dif = data['ECEF-Y'].diff()
    alt_dif = data['ECEF-Z'].diff()
    
    data['Distance'] = np.sqrt(lat_dif**2 + lon_dif**2 + alt_dif**2)
    
    return data

# creating the dataframes with different magnitude filters
different_m_distance_list = []
for magnitude in magnitudes:
    different_m_distance_list.append(find_distance(data_distance_mag, magnitude))
```

#### Analyzing scaling law for waiting times

Here, we observed a meaningful pattern in average meaning time. Higher magnitude means higher waiting times.

```python
# loading the data
data = pd.read_pickle("data.pkl")

# getting magnitude ranges to be able to analyse if there is a meaningful pattern
magnitude_ranges = list(np.arange(data['Magnitude'].min(),data['Magnitude'].max() - 1, 0.1)) # Set the values of m

def find_waiting_time_mean(data, m):
    data = data[data['Magnitude'] >= m]
    data['diff'] = data['Second'].diff()
    data = data.dropna()
    mean = data['diff'].mean()
    
    return mean

waiting_time_means = []
for manitude in magnitude_ranges: # Compute the average waiting time after filtering for given m
    waiting_time_means.append(find_waiting_time_mean(data, manitude))
    
fig = plt.subplots(figsize=(10,8))
plt.plot(magnitude_ranges, waiting_time_means, label="data", markersize=10, c="black", marker="o")
plt.xscale('log')
plt.yscale('log')
plt.title('Average waiting time with different magnitude values in log-log scale')
plt.xlabel('Magnitude')
plt.ylabel('Average waiting time [s]')
plt.show()
```

Since the data have a power-law degree distribution, our data possess no characteristic scale which means there is not any typical point in the data that represents the degree for the other points. In this situation, we scaled our data with the mean of the waiting time. So the shape of the data will be the same and the visualization of data frames that are filtered by different values of m will collapse on the same curve.

```python
# loading the data
data = pd.read_pickle("data.pkl")

# magnitude list
magnitudes = [2, 3, 4, 5, 6]

# the function that creates dataframes that include waiting time based on the given magnitude.
def find_waiting_time(data, m):
    data = data[data['Magnitude'] >= m]
    data['WaitingTime'] = data['Second'].diff()
    return data

# creating a list that holds waiting times filtered by certain magnitudes
data_list = []
for i in range(len(magnitudes)):
    data_list.append(find_waiting_time(data, magnitudes[i]))
    data_list[i].dropna()

# parameters to hold values related with "WaitingTime" in the dataframes above.
x_list_waiting_time = [] # bin centers
y_list_waiting_time = [] # frequency

for i in range(len(magnitudes)):
    # creating widths
    width = 1000*np.logspace(0, 3, int(np.sqrt(data.shape[0]))+1)
    
    # getting the edges (bins)
    edge = np.cumsum(width)
    
    # getting the h, bin, patches of the dataframes.
    fig, ax = plt.subplots(figsize=(10,8))
    ax = sns.distplot(data_list[i]['WaitingTime'], bins = edge, kde = False, norm_hist = True)
    
    # edges from hist (left border)
    edge = np.array([h.get_xy()[0] for h in ax.patches])
    
    # widths from hist
    width = np.array([h.get_width() for h in ax.patches])
    
    # bin centers
    x_list_waiting_time.append(np.add(edge, width/2))
    
    # frequency
    y_list_waiting_time.append(np.array([h.get_height() for h in ax.patches]))
    
    # hide the plotting
    plt.close()

# finding the means of waiting times
waiting_time_mag_means = []
for i in range(len(magnitudes)):
    waiting_time_mag_means.append(data_list[i]['WaitingTime'].mean()) 

    
x_waiting_time_scaled = [0, 0, 0, 0, 0]
y_waiting_time_scaled = [0, 0, 0, 0, 0]
for i in range(len(magnitudes)):
    x_waiting_time_scaled[i] = x_list_waiting_time[i]/waiting_time_mag_means[i]
    y_waiting_time_scaled[i] = y_list_waiting_time[i]*waiting_time_mag_means[i]

# original data    
fig, ax = plt.subplots(1,figsize=(16,8))

# drawing each original waiting time filtered by magnitude
for i in range(len(magnitudes)):
    ax.plot(x_list_waiting_time[i], y_list_waiting_time[i],'.', label = ("m = " + str(magnitudes[i])))
    
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_title('Original Data')
ax.set_xlabel('Waiting time [s]')
ax.set_ylabel('Normalized Freq.')
ax.legend()
plt.show()

# scaled data
fig, ax = plt.subplots(1,figsize=(16,8))

# drawing each scaled waiting time filtered by magnitude
for i in range(len(magnitudes)):
    ax.plot(x_waiting_time_scaled[i],y_waiting_time_scaled[i],'.', label = ("m = " + str(magnitudes[i])))
    
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_title('Scaled by mean of waiting time')
ax.set_xlabel('Waiting time [s]')
ax.set_ylabel('Normalized Freq.')
ax.legend()
plt.show()    
```
